```python
%matplotlib inline

import matplotlib.pyplot as plt
from IPython.display import display as d
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(12,12)})
sns.set_style("white")
import glob
import json
import os
import sys
import shutil
```

```python
df = pd.read_csv("si_src.csv")
d(df)
```

```python
def save_fig(ax, title, pdfname):
    ax.set_title(title)
    ax.get_figure().savefig(pdfname, bbox_inches="tight")
```

```python
sns.set(rc={'figure.figsize':(8,1)})
sns.set_style("white")

ax = sns.boxplot(x="si", data=df)
ax.set_xlabel("SI")
save_fig(ax, "SI for source images.", "si_src.pdf")
```
