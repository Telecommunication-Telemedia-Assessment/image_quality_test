```python
%matplotlib inline

import matplotlib.pyplot as plt
from IPython.display import display as d
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(8,1)})
sns.set_style("white")
import glob
import json
import os
import sys
import shutil
```

```python
def save_fig(ax, title, pdfname):
    ax.set_title(title)
    ax.get_figure().savefig(pdfname, bbox_inches="tight")
```

```python
df = pd.read_json("vca_src.json.gz")
df["Spatial complexity"] = df["E"]
d(df)
```

```python
sns.set(rc={'figure.figsize':(8,1)})
sns.set_style("white")

ax = sns.boxplot(x="E", data=df, color="g")
ax.set_xlabel("E")
save_fig(ax, "Spatial complexity (VCA) for all encoded images.", "sc_encoded.pdf")
```

```python
# averageBrightness
ax = sns.boxplot(x="averageBrightness", data=df, color="m")
ax.set_xlabel("Brightness")
save_fig(ax, "Average Brightness (VCA) for all encoded images.", "bright_encoded.pdf")
```

```python
ds = pd.read_json("si_encoded.json.gz")
d(ds)
```

```python
ax = sns.boxplot(x="si", data=ds, color="b")
ax.set_xlabel("SI")
save_fig(ax, "SI for all encoded images.", "si_encoded.pdf")
```
