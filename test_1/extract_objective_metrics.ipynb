```python
reports = list(glob.glob("reports/*/*.json"))
print(reports[0:2])

src_images = """air_acrobatics_harmonic.mkv_1frame.mkv
american_football_harmonic.mkv_1frame.mkv
animals_harmonic.mkv_1frame.mkv
asian_fusion_harmonic.mkv_1frame.mkv
BennuProRes4444.mov_1frame.mkv
BuildingHall2_3840x2160_50fps_10bit_420_ffvhuff.mkv_1frame.mkv
Campfire_3840x2160_30fps_bt709_420_videoRange_ffvhuff.mkv_1frame.mkv
CatRobot_3840x2160_60fps_10bit_420_jvet_ffvhuff.mkv_1frame.mkv
classroom.mkv_1frame.mkv
costa_rica_harmonic.mkv_1frame.mkv
Dancers.avi_1frame.mkv
DaylightRoad2_3840x2160_60fps_10bit_420_ffvhuff.mkv_1frame.mkv
Drums_3840x2160_100fps_10bit_420_jvet_ffvhuff.mkv_1frame.mkv
ein_abend_zu_zweit_ffvhuff.mkv_1frame.mkv
epicdragon-skintone-6k-ws-96fps_444.mkv_1frame.mkv
gooseberry.mkv_1frame.mkv
hong_kong_harmonic.mkv_1frame.mkv
monkeys_harmonic.mkv_1frame.mkv
myanmar_harmonic.mkv_1frame.mkv
Netflix_BarScene_4096x2160_60fps_10bit_420.y4m_1frame.mkv
Netflix_DinnerScene_4096x2160_60fps_10bit_420.y4m_1frame.mkv
Netflix_DrivingPOV_4096x2160_60fps_10bit_420.y4m_1frame.mkv
NETFLIX_ElFuente_4096x2160_28592frames_5994fps.mov_1frame.mkv
Netflix_PierSeaside_4096x2160_60fps_10bit_420.y4m_1frame.mkv
NTIA_4K_Vignette_1_Violin-422p-8bit-60fps-4096x2160.mkv_1frame.mkv
rally_harmonic.mkv_1frame.mkv
raptors_harmonic.mkv_1frame.mkv
raven-45k-48fps_444.mkv_1frame.mkv
red_rocks_harmonic.mkv_1frame.mkv
RollerCoaster_4096x2160_60fps_10bit_420_jvet_ffvhuff.mkv_1frame.mkv
Sparks_20170214.mxf_1frame.mkv
Tango_4096x2160_60fps_10bit_420_jvet_ffvhuff.mkv_1frame.mkv
ToddlerFountain_4096x2160_60fps_10bit_420_jvet_ffvhuff.mkv_1frame.mkv
TrafficFlow_3840x2160_30fps_10bit_420_jvet_ffvhuff.mkv_1frame.mkv
venice_harmonic.mkv_1frame.mkv
weapon8k-standard-60fps-12to1redcode_16x9_444.mkv_1frame.mkv
weapon-h2o-6k-24fps_444.mkv_1frame.mkv
weapon-standard-6k-ff-24fps_444.mkv_1frame.mkv""".split("\n")

src_images = [os.path.splitext(os.path.splitext(x)[0])[0] for x in src_images]
print("\n".join(src_images))

def read_report(x):
    with open(x) as xfp:
        j = json.load(xfp)
    x = os.path.splitext(os.path.basename(x))[0]
    
    for y in src_images:
        x = x.replace(y + "_", "", 1)
        pass
    x = x.replace("encoded_", "")
    
    return {
        "image": x,
        "vmaf": j["frames"][0]["metrics"]["vmaf"],
        "adm2": j["frames"][0]["metrics"]["adm2"],
        "psnr": j["frames"][0]["metrics"]["psnr"],
        "ssim": j["frames"][0]["metrics"]["ssim"],        
        "ms_ssim": j["frames"][0]["metrics"]["ms_ssim"],  
        "vif_scale0": j["frames"][0]["metrics"]["vif_scale0"],
        "vif_scale1": j["frames"][0]["metrics"]["vif_scale1"],
        "vif_scale2": j["frames"][0]["metrics"]["vif_scale2"],
        "vif_scale3": j["frames"][0]["metrics"]["vif_scale3"]
    }

values = list(map(read_report, reports))

df = pd.DataFrame(values)


d(df.head())
print(len(df))

```
